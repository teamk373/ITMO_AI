Создаем шоколадку

```python
class Chocolate:
  def __init__ (self, price:float, have_nuts:bool, nuts_count:int, addition:str = None):
    if (price > 0):
      self.price = price
    else:
      raise ValueError('Цена должна быть положительной')
    self.addition = addition
    self.have_nuts = have_nuts
    self.nuts_count = nuts_count

  def show_my_chocolate(self):
    if str != None:
      try:
        print("Шоколадный батончик, добавка: "+self.addition+'\n')
      except:
        print("Обычный шоколадный батончик"+'\n')


  def count_nuts(self):
    for i in range (self.nuts_count):
      print("*", end='')
    print()

  def __str__(self):
    return "Шоколадный батончик, "+str(self.price)+" рублей"
```

```python
choco1 = Chocolate(100.1, True, 5)
print(choco1)
choco1.count_nuts()
choco1.show_my_chocolate()
choco2 = Chocolate(20, False, 0, "Nougat")
choco2.show_my_chocolate()
choco3 = Chocolate(10, True, 1, "Cream")
print(choco3)
choco3.count_nuts()
choco3.show_my_chocolate()
choco4 = Chocolate(0,False,0)
```
