Подключение датасета и вывод первых строк

```python
import pandas as pd
import re

df = pd.read_csv("train.csv")
df.head(30)
```

Вывод основной информации о датасете

```python
print("\n Информация о типах данных")
df.info()
print("\n Число пропусков в каждом столбце")
print(df.isnull().sum())
print("\n Статистическое описание")
df.describe()
```

Процент выживаемости у каждого класса

```python
survival_percent = df.groupby('Pclass')['Survived'].mean() * 100
print(survival_percent)
```

Функция для выделения имени и вывод самых популярных имеш

```python
def extract_first_name(fullname):
    bracket_name = re.search(r"\((.*?)\)", fullname)
    if bracket_name:
        return bracket_name.group(1).split()[0]
    match = re.search(r", [^\.]+\. ([^\(]+)", fullname)
    return match.group(1).split()[0] if match else None

df['FirstName'] = df['Name'].apply(extract_first_name)

male_name = df[df['Sex'] == 'male']['FirstName'].mode()[0]
female_name = df[df['Sex'] == 'female']['FirstName'].mode()[0]

print("Самое популярное мужское:", male_name)
print("Самое популярное женское:", female_name)
```

Самое популярное мужское имя по классам

```python
df[df['Sex']=='male'].groupby('Pclass')['FirstName'].agg(lambda x: x.mode()[0])
```

Самые популярные женские имена по классам

```python
df[df['Sex']=='female'].groupby('Pclass')['FirstName'].agg(lambda x: x.mode()[0])
```
